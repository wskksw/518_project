```python
import matplotlib.pyplot as plt

features = ["ca, thal", "ca", "ca, thal, cp", "ca, thal, oldpeak, cp"]
accuracies = [0.86, 0.83, 0.80, 0.83]

plt.figure(figsize=(10, 6))
plt.bar(features, accuracies, color='skyblue')
plt.xlabel('Feature Combinations')
plt.ylabel('Accuracy')
plt.title('Iterations of Feature Selection (Binary Classification)')
plt.ylim(0.0, 1)  


plt.show()
```

```python

metrics = ['Accuracy', 'Precision', 'Average Recall', 'F1 Score']
logistic_regression_values = [0.83, 0.91, 0.71, 0.8]
random_forest_values = [0.87, 0.86, 0.86, 0.86]


x = range(len(metrics))  
width = 0.35  
fig, ax = plt.subplots(figsize=(10, 4))
bars1 = ax.bar(x, logistic_regression_values, width, label='Logistic Regression', color='skyblue')
bars2 = ax.bar([p + width for p in x], random_forest_values, width, label='Random Forest', color='lightgreen')

ax.set_xlabel('Metrics')
ax.set_ylabel('Values')
ax.set_title('Binary Classification Model Comparison')
ax.set_xticks([p + width / 2 for p in x])
ax.set_xticklabels(metrics)

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()

```
