```python
import pandas as pd

# Convert features and targets to DataFrame for better visualization and manipulation
df_features = pd.DataFrame(X)
df_targets = pd.DataFrame(y)

# Display basic information
print(df_features.info())

# Display basic statistics for each feature
print(df_features.describe())


import matplotlib.pyplot as plt
import seaborn as sns

# Distribution plots for each feature
for column in df_features.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df_features[column], bins=30, kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()

# Distribution of target variable
plt.figure(figsize=(10, 5))
sns.countplot(x="num", data=df_targets)
plt.title('Distribution of Heart Disease')
plt.show()

# Correlation heatmap
correlation_matrix = df_features.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation between features')
plt.show()


# Check for missing values in the features
missing_data = df_features.isnull().sum()
missing_percentage = (df_features.isnull().sum() / len(df_features)) * 100

print("Missing Data Count:\n", missing_data)
print("\nMissing Data Percentage:\n", missing_percentage)





```
